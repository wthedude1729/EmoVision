```python
import cv2

# the following was stolen from StackOverflow
def get_frames(readpath, writepath):
    cap = cv2.VideoCapture(readpath)
    i = 0
    # a variable to set how many frames you want to skip
    frame_skip = 72 # every 3 seconds, we make another frame
    # a variable to keep track of the frame to be saved
    frame_count = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        if i > frame_skip - 1:
            frame_count += 1
            cv2.imwrite(writepath + str(frame_count) + ".jpg", frame)
            i = 0
            continue
        i += 1
    cap.release()
    cv2.destroyAllWindows()
```

```python
directory = "/content/drive/My Drive/Machine Learning/COSMOS/FINAL_PROJECT/DER/ekman6/"
to_directory = "/content/drive/My Drive/Machine Learning/COSMOS/FINAL_PROJECT/DER/ekman6_frames/"
folders = ["anger/", "disgust/", "fear/", "joy/", "sadness/", "surprise/"]

for i in range(6):
  path = directory + folders[i]
  for file in os.listdir(path):
    filename = os.fsdecode(file)
    write_to = to_directory + folders[i] + filename + "/"
    get_frames(path + filename, write_to)
```
